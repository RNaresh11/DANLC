Lab 1 .Analyze the relationship between the size of houses (measured in square
footage) and their selling prices in a particular neighborhood. You have collected
data on various houses in that neighborhood.Create a scatter plot using the
below data and share your conclusion/analysis

```python
square_footage = np.array([1200, 1400, 1600, 1800, 2000, 2200, 2400, 2600, 2800,3000])
selling_prices = np.array([250, 290, 315, 380, 410, 450, 500, 525, 570, 610])
```

```python
plt.scatter(square_footage,selling_prices,marker="o",color="green")
plt.xlabel("Square Footage (sq.ft)")
plt.ylabel("Selling Price ($000s)")
plt.title("Housing Prices vs. Square Footage",c='g',fontweight=30)
plt.grid()
plt.show()
```

Lab2: Create a pie chart to visualize the distribution of your monthly income by
source. You have collected data on the various sources of your income, such as
salary, freelance work, investments, and rental income. Share your
conclusion/analysi

```python
income_sources = ['Salary', 'Freelance', 'Investments', 'Rental', 'Other']
monthly_income = [5000, 1500, 1000, 600, 400]
```

```python
def format_pct(pct):
    total = sum(monthly_income)
    value = int(pct*total/100)
    return f'{pct:.1f}%\n({value})'


plt.pie(monthly_income,labels=income_sources,autopct=format_pct,counterclock=False,radius=1.5)
plt.show()
```

Lab3: Create a pie chart to illustrate the distribution of a company's revenue
 across its various business segments. You have collected data on the revenue
 generated by each segment, such as Product A, Product B, Services, and
 Licensing. Share your conclusion/analysis.

```python
segments = ['Product A', 'Product B', 'Services', 'Licensing']
revenue_percentages = [45, 25, 15, 15]
```

```python
def format_pct(pct):
    total = sum(revenue_percentages)
    value  = int(pct*total/100)
    return f'{pct:.1f}%\n({value})'

plt.pie(revenue_percentages,labels=segments,autopct=format_pct,counterclock=False)
plt.show()
```

 Lab3: Suppose you're a sales manager for an e-commerce company, and you
 want to create a figure with subplots to compare the sales performance of
 different product categories over time. You have sales data for four product
categories: Electronics, Clothing, Home & Garden, and Sports & Outdoors. Share
 your conclusion/analysis.

```python
months = np.arange(1, 13)
electronics_sales = np.array([25000, 28000, 31000, 27000, 30000, 32000, 35000,36000, 38000, 39000, 41000, 42000])
clothing_sales = np.array([15000, 16000, 17000, 18000, 19000, 20000, 21000, 22000,23000, 24000, 25000, 26000])
home_garden_sales = np.array([18000, 19000, 20000, 21000, 22000, 23000, 24000,25000, 26000, 27000, 28000, 29000])
sports_outdoors_sales = np.array([12000, 13000, 14000, 15000, 16000, 17000, 18000,19000, 20000, 21000, 22000, 23000])
```

```python
fig,axs = plt.subplots(2,2,figsize=(12,10))

axs[0,0].plot(months,electronics_sales,c="blue",marker="o")
axs[0,0].set_title("Elecctronics")
axs[0,0].set_xlabel('Sales Amount')
axs[0,0].set_ylabel('Sales')

axs[0,1].plot(months,clothing_sales,c="g",marker="o")
axs[0,1].set_title("Clothing Sales")
axs[0,1].set_xlabel('Sales Amount')
axs[0,1].set_ylabel('Sales')

axs[1,0].plot(months,home_garden_sales,c="r",marker="o")
axs[1,0].set_title("Home Gardern Sales")
axs[1,0].set_xlabel('Sales Amount')
axs[1,0].set_ylabel('Sales')

axs[1,1].plot(months,sports_outdoors_sales,c="magenta",marker="o")
axs[1,1].set_title("Sports Outdoor Sales")
axs[1,1].set_xlabel('Sales Amount')
axs[1,1].set_ylabel('Sales')
```
